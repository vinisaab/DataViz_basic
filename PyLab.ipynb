# PyLab

O PyLab combina funcionalidades do NumPy com o MatPlotLib

```python
#importando modulos
from pylab import *
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## Grafico de Linha

```python
x = linspace(0,5,10)
y = x ** 2

fig = plt.figure()

#definindo os eixos
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('grafico de Linha')
```

## Multiplos Graficos

### grafico dentro de grafico

```python

fig = plt.figure()

#definindo os eixos
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('grafico de Linha')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(x,y,'g')
axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('grafico de Linha2')
```

### graficos em paralelo

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x,y,'r')
    
fig.tight_layout()
```
